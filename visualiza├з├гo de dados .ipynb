### Vamos entender qual a relação da idade com o fatoe de sobrevivência no Titanic

```python
# Grafico básico - linhas 

plt.plot(base.age)
```

```python
# Grafico de Barras

plt.bar(base.survived, base.age.mean())
```

```python
#grafico de barras Horizontais 
plt.barh(base.survived, base.age.mean())
```

```python
# grafico de dispersão
plt.scatter(base.survived, base.age)
```

```python
# Removendo valores que saõ nulos 

base2 = base.dropna(axis = 0, subset = ['age'])
```

```python
plt.hist(base2.age)
```

# Montando a Tela

```python
plt.figure(figsize=(20, 10))
```

# Editando a tela 


```python
plt.figure(figsize=(20, 10))

plt.title('Histograma de idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
```

# Plotando o grafico

```python
plt.figure(figsize=(20, 10))
plt.title("Histograma da idade dos Passageiros ")
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.hist(base2.age)
```

# Mudando a cor Do Grafico

```python
plt.figure(figsize=(20, 10))
plt.title('Histograma de idade dos passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequencia')

plt.hist(base2.age, color='green')
```

# Separando os graficos( um para sobrevivente e outro para não sobrevivente

```python
plt.figure(figsize=(20, 10))
plt.title('Histograma de idade dos passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.hist(base2[base2.survived == 0].age, color = 'red')
```

```python
plt.figure(figsize=(20,10))
plt.title('Histograma de Idade dos passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.hist(base2[base2.survived==1].age, color='blue')
```

```python
plt.figure(figsize=(20, 10))
plt.title('Histograma de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.hist(base2[base2.survived==0].age, color= 'red')
plt.hist(base2[base2.survived==1].age, color='blue')
```
