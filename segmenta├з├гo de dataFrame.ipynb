# segmentação de DataFames(slicing)

## metodos usados para segmentação de DataFrames:

#### Loc = Usado para segmentação usando rótulos de indices e colunas => df.loc[ Linhas, colunas]

#### ILoc = Usado para segmentação usando números referentes à indices e colunas => df.iloc[linhas, colunas]

#### Segmentação por índices => df[indices]

#### Segmentação por colunas => df['colunas'] ou df[['coluna1' , 'coluna2']]
