#  Dica -  função Apply

```python
def funcaoNome(nome):
    """Essa função recebe um argumento, e pega a primeiro e altima letra e acrecenta a palavra 'final' """
    return nome[0]+nome[-1] + '  Final'
```

```python
base['colunaNovaFunçãoApply'] = base['Name'].apply(lambda x: funcaoNome(x))
```

```python
base
```

```python
base['PrimeiroNome'] = base['Name'].apply(lambda priNome: priNome.split()[0])
# estou criando uma coluna chamada primeiroNome e essa coluna esta recebendo os dados da coluna Name e usa a função apply para 
# aplicar uma funçao lambda que usa o metodo splir que tranforma cada linha da coluna Name em uma lista e pega apenas o elemento
# na plosição 0
```
