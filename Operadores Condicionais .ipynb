```python
def funcaoNome(nome):
    return nome[0]+nome[-1] + '  Final'
```

```python
base['colunaNova'] = pd.Series()
```

```python
for i in base.index:
    if base['Name'][i][0] == "K":
        base['colunaNova'][i] = funcaoNome(base['Name'][i])
    elif base['Name'][i][0] ==  "W":
        base['colunaNova'][i] = funcaoNome(base['Name'][i])
    else:
        pass
```
